```python
'''Definir uma função average_spam_confidence que calcula a confiança média de spam e a retorna
Abra o arquivo mbox-short.txt que está localizado em /cxldata/datasets/project/mbox-short.txt
Faça loop através do identificador do arquivo
Selecione apenas as linhas começam com X-DSPAM-Confidence:
Divida as linhas em : e pegue o valor de flutuação que é a confiança de spam
Encontre a média dessa confiança de spam em todo o arquivo e retorne-o.
PS - Se a sua lógica estiver correta, então a pontuação correta de confiança de spam deve ser 0.7507185185185187.'''
```

```python
def average_spam_confidence(file_path):
    with open(file_path, 'r') as file:
        spam_confidence = [float(line.split(':')[-1]) for line in file if line.startswith('X-DSPAM-Confidence:')]
    if spam_confidence:
        return sum(spam_confidence) / len(spam_confidence)
    else:
        return 0
average_spam_confidence('mbox-short.txt')
```

```python
'''Encontre Qual Dia da Semana o E-mail foi enviado
Escreva uma função find_email_sent_days que lê o arquivo /cxldata/datasets/project/mbox-short.txt e categoriza cada mensagem de correio em que dia da semana o e-mail foi enviado.

Para fazer isso, faça o seguinte:

Abra o arquivo e leia-o linha por linha
Procure por linhas que comecem com "De"
Para as linhas que começam a partir de "De", então procure o terceira palavra e manter uma corrida contar de cada um dos dias da semana. Como você encontra o dia da semana, é um exercício para você.
> Nota: Você tem que armazenar os resultados em um dictionary. Armazene apenas aqueles dias da semana que existem. Por exemplo, se não houver linha para Mon então não deve estar nos elementos do dicionário.

No final do programa, devolva o conteúdo do seu dictionary (a ordem não importa)
Linhas de amostra do arquivo:

From stephen.marquard@uct.ac.za Sat Jan  5 10:14:16 2008
From stephen.marquard@uct.ac.za Sat Jan  5 15:14:16 2008
From stephen.marquard@uct.ac.za Sun Jan  6 09:14:16 2008
Saída:

{'Sat': 2, 'Sun': 1}
PS - Se a sua lógica estiver correta, então a sua função deve retornar este dicionário {'Sat': 1, 'Fri': 20, 'Thu': 6}. '''
```

```python
def find_email_sent_days(file_path):
    days_count = {}  
    with open(file_path, 'r') as file:
        for line in file:
            if line.startswith('From '): 
                words = line.split()  
                if len(words) > 2: 
                    day = words[2] 
                    days_count[day] = days_count.get(day, 0) + 1 
    return {key: value for key, value in days_count.items() if value > 0} 
find_email_sent_days('mbox-short.txt')
```
